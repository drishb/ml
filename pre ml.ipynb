```python
start_time=tm.time()
matrix_1=[
    [1,2,3],
    [4,5,6]]
matrix_2=[
    [1,1],
    [1,1],
    [1,1]]
sol_matrix=[
    [0,0],
    [0,0]]

for i in range(len(matrix_1)):
        for j in range(len(matrix_2[0])):
            for k in range(len(matrix_2)):
               sol_matrix[i][j]= sol_matrix[i][j]+matrix_1[i][k]*matrix_2[k][j]
end_time=tm.time()
                
            
```

```python
print(sol_matrix, end_time-start_time)
```
